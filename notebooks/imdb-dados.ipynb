```python
import pandas as pd

nome_real_do_arquivo = 'title.basics.tsv' 
caminho_do_arquivo = f'../data/raw/{nome_real_do_arquivo}'

df = pd.read_csv(caminho_do_arquivo, sep='\t', low_memory=False) 

df.head()
```

```python
print(f"O conjunto de dados possui {df.shape[0]} amostras (linhas) e {df.shape[1]} atributos (colunas).") 
print("\n" + "="*30 + "\n")

print("1. Tipos de Dados e Contagem de Não-Nulos:")
df.info()

print("\n" + "="*30 + "\n")

print("2. Estatísticas Descritivas:")
print(df.describe())

print("\n" + "="*30 + "\n")

print("3. Contagem de Valores Faltantes (df.isna().sum()):")
print(df.isna().sum())
```
